```python
import copy
```

```python
def printBoard(board):
    for row in board:
        print(row)
    print()
```

```python
def score (b):

    player = '0'
    opponent = 'X'

    for row in range(3) :      
        if (b[row][0] == b[row][1] and b[row][1] == b[row][2]) :         
            if (b[row][0] == player) : 
                return 1
            elif (b[row][0] == opponent) : 
                return -1

    for col in range(3) : 
       
        if (b[0][col] == b[1][col] and b[1][col] == b[2][col]) : 
          
            if (b[0][col] == player) :  
                return 1
            elif (b[0][col] == opponent) : 
                return -1
  
    if (b[0][0] == b[1][1] and b[1][1] == b[2][2]) : 
      
        if (b[0][0] == player) : 
            return 1
        elif (b[0][0] == opponent) : 
            return -1
  
    if (b[0][2] == b[1][1] and b[1][1] == b[2][0]) : 
      
        if (b[0][2] == player) : 
            return 1
        elif (b[0][2] == opponent) : 
            return -1
  
    return 0
```

```python
def generateTree( board, move, minmax):

    n = len(board)
    # printBoard(board)

    emptyCell = 0

    for i in range(n):
        for j in range(n):
            if board[i][j] == ' ':
                emptyCell += 1
    
    if emptyCell == 0:
        s = score(board)
        # print(s)
        return board,s

    newMove = copy.deepcopy(move)
    if move == '0':
        newMove = 'X'
    else:
        newMove = '0'

    bestScore = 0
    bestBoard = copy.deepcopy(board)
    
    newMinMax = copy.deepcopy(minmax)
    if minmax == 'Min':
        bestScore = 10
        newMinMax = 'Max'
    else:
        bestScore = -10
        newMinMax = 'Min'

    for i in range(n):
        for j in range(n):

            newBoard = copy.deepcopy(board)

            if board[i][j] == ' ':
                newBoard[i][j] = move
                curBoard,curScore = generateTree(newBoard, newMove, newMinMax)

                if minmax == "Min":
                    if curScore < bestScore:
                        bestBoard = copy.deepcopy(curBoard)
                        bestScore = copy.deepcopy(curScore)

                if minmax == "Max":
                    if curScore > bestScore:
                        bestBoard = copy.deepcopy(curBoard)
                        bestScore = copy.deepcopy(curScore)

    print(bestScore)
    printBoard(bestBoard)
    return bestBoard,bestScore
```

```python
board=[
    [' ',' ',' '],
    [' ',' ',' '],
    [' ',' ',' '],
]
move ='0'

# printBoard(board)
board,score = generateTree( board, move, "Min")
print(score)
printBoard(board)
```
