```python
supply = {
    "VLS" : 0,
    "LS" : 20,
    "MS" : 40,
    "HS" : 60,
    "VHS" : 80
}

def mf_supply(x):
    if x<0:
        return [
            {"val":(20-x)/20,"s":"VLS",},
            {"val":x/20,"s":"LS"}
        ]
    if x<20:
        val1 = (20-x)/(20-0)
        s1 = "VLS"
        val2 = 1-val1
        s2 = "LS"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x<40:
        val1 = (40-x)/(40-20)
        s1 = "LS"
        val2 = 1-val1
        s2 = "MS"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x<60:
        val1 = (60-x)/(60-40)
        s1 = "MS"
        val2 = 1-val1
        s2 = "HS"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x <= 80:
        val1 = (80-x)/(80-60)
        s1 = "HS"
        val2 = 1-val1
        s2 = "VHS"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    else:
        return  [{"val":(100-x)/20,"s":"VHS"}]

demand = {
    "LD":25,
    "MD":50,
    "HD":75
}

def mf_demand(x):
    if x<25:
        return  [{"val":1+(25-x)/15, "s":"LD"}]
    if x<50:
        val1 = (50-x)/(50-25)
        s1 = "LD"
        val2 = 1-val1
        s2 = "MD"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x<=75:
        val1 = (75-x)/(75-50)
        s1 = "MD"
        val2 = 1-val1
        s2 = "HD"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]

price ={
    "VLPG":10,
    "LPG":25,
    "MPG":45,
    "HPG":60,
    "VHPG":80
}

def mf_price(x):
    if x<10:
        return [{"val":(10-x)/10, "s":"VLPG"}]
    if x<25:
        val1 = (25-x)/(25-10)
        s1 = "VLPS"
        val2 = 1-val1
        s2 = "LPG"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x<45:
        val1 = (45-x)/(45-25)
        s1 = "LPG"
        val2 = 1-val1
        s2 = "MPG"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x<60:
        val1 = (60-x)/(60-40)
        s1 = "MPG"
        val2 = 1-val1
        s2 = "HPG"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    if x < 80:
        val1 = (80-x)/(80-60)
        s1 = "HPG"
        val2 = 1-val1
        s2 = "VHPG"
        return [
            {"val":val1,
            "s":s1,},
            {"val":val2,
            "s":s2}
        ]
    else:
        return  [{"val":1+x/10, "s":"VHPG"}]
```

```python
def rule( sup, dem):
    if sup == "VLS" and dem == "LD":
        return "LPG"
    if sup == "VLS" and dem == "MD":
        return "HPG"
    if sup == "VLS" and dem == "HD":
        return "VHPG"
    
    if sup == "LS" and dem == "LD":
        return "LPG"
    if sup == "LS" and dem == "MD":
        return "HPG"
    if sup == "LS" and dem == "HD":
        return "VHPG"
    
    if sup == "MS" and dem == "LD":
        return "VLPG"
    if sup == "MS" and dem == "MD":
        return "MPG"
    if sup == "MS" and dem == "HD":
        return "HPG"
    
    if sup == "HS" and dem == "LD":
        return "VLPG"
    if sup == "HS" and dem == "MD":
        return "MPG"
    if sup == "HS" and dem == "HD":
        return "MPG"
    
    if sup == "VHS" and dem == "LD":
        return "VLPG"
    if sup == "VHS" and dem == "MD":
        return "LPG"
    if sup == "VHS" and dem == "HD":
        return "MPG"
```

```python
def defuzzify(sym, val):
    if sym=="VLPG":
        return (25-(val*15))  
    if sym=="LPG": 
        return (55-(val*5)/2)
    if sym=="MPG": 
        return (85+val*5)/2 
    elif sym=="HPG": 
        return (125-val*5)/2 
    else: 
        return (90-(val*10)) 
```

```python
input_supply = 50
input_demand = 60

fuz_supply = mf_supply(input_supply)
fuz_demand = mf_demand(input_demand)

print(fuz_supply)
print(fuz_demand)

fuz_price = []

for sup in fuz_supply:
    for dem in fuz_demand:
        # print(sup)
        # print(dem)
        sym = rule(sup["s"], dem["s"])
        val = min(sup["val"], dem["val"])
        fuz_price.append({'val': val, "s":sym})

print(fuz_price)

fuz_ans = fuz_price[0]

for data in fuz_price:
    if( data['val'] > fuz_ans['val']):
        fuz_ans = data

print(fuz_ans)

ans = defuzzify(fuz_ans["s"], fuz_ans["val"])

print('supply = ',input_supply)
print('demand = ',input_demand)

print('price = ',ans)
```
